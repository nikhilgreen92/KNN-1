```python
data['GENDER'] = pd.get_dummies(data['GENDER'], drop_first=True) # drop_first=True is used so as to avoid dummy variable trap.
```

```python
data['Car_Owner'] = pd.get_dummies(data['Car_Owner'], drop_first=True) # drop_first=True is used so as to avoid dummy variable trap.
```

```python
data['Propert_Owner'] = pd.get_dummies(data['Propert_Owner'], drop_first=True) # drop_first=True is used so as to avoid dummy variable trap.
```

```python
data['Type_Income'] = data.Type_Income.map({'Working': 0, 'Commercial associate': 1, 'Pensioner': 2, 'State servant': 3,}) # Mapping
data.head()
```

```python
data['EDUCATION'] = data.EDUCATION.map({'Secondary / secondary special': 0, 'Higher education': 1, 'Incomplete higher': 2, 'Lower secondary': 3, 'Academic degree' : 4}) # Mapping
data.head()
```

```python
data['Marital_status'] = data.Marital_status.map({'Married': 0, 'Single / not married': 1, 'Civil marriage': 2, 'Separated': 3, 'Widow' : 4}) # Mapping
data.head()
```

```python
data['Housing_type'] = data.Housing_type.map({'House / apartment': 0, 'With parents': 1, 'Municipal apartment': 2, 'Rented apartment': 3, 'Office apartment' : 4, 'Co-op apartment' : 5}) # Mapping
data.head()
```

```python
data.dtypes # Checking the dtypes of all columns
```

```python
# converting required column types into categorical datatype
for i in data.columns:
  if (i == 'Ind_ID' or i == 'CHILDREN' or i == 'Annual_income' or i == 'Birthday_count' or i == 'Employed_days'):
    pass
  else:
    data[i] = data[i].astype('category')
```

Now we are going to scale the data but before that we are going to split the data into training and testing data.

```python
# Splitting the data into X(independent columns) and y(dependent column)
from sklearn.model_selection import train_test_split
X = data[['GENDER','Car_Owner','Propert_Owner','CHILDREN','Annual_income','Type_Income','EDUCATION','Marital_status','Housing_type','Mobile_phone',	'Work_Phone',	'Phone',	'EMAIL_ID','Family_Members']]
y = data['label']
```

```python
# Splitting the data into training and testing set.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state = 42)
```

```python
# Scaling the data using standard scaler.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Using K-NN algorithm.

```python
# Importing k neighbours classifier.
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(X_train,y_train)
```

```python
# Predicting y values using KNN Algorithm algorithm.
y_pred = classifier.predict(X_test)
```

# Evaluation of the model

```python
# Importing metrics to calculate accuracy, precision, recall, f1 score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
 # Calculating accuracy of y_values
accuracy_knn = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy_knn*100) # calculating accuracy in percentage terms
```

Since we are getting an accuracy of 89% we will use use hyperparameters to tune the values so that accuracy score is increased.

```python
# Importing gridsearch cv to hypertune parameters.
from sklearn.model_selection import GridSearchCV
```

```python
# Hyperparametric tuning
estimator = KNeighborsClassifier()
param_grid = {'n_neighbors' : [2,3,4,5,6,7,8,9,10],
               'weights' : ['uniform','distance']}
grid_search = GridSearchCV(estimator, param_grid, cv = 5, scoring = 'accuracy')
```

```python
# fitting the model for grid search
grid_search.fit(X_train,y_train)
```

```python
print(grid_search.best_params_)
```

```python
# training with best hyperparameters as we got in previous result just above
better_knn = KNeighborsClassifier(n_neighbors = 10, weights = 'distance')
better_knn.fit(X_train, y_train)
y_pred_knn = better_knn.predict(X_test) # Predicting model using KNN Algorithm algorithm.
```

```python
accuracy_knn = accuracy_score(y_test,y_pred_knn)
print('Test accuracy:', accuracy_knn*100) # calculating accuracy
```

We do see here that our accuracy for the model has incresed by 3% and it's 92% now.

```python
# confusion matrix
con_mat_knn = confusion_matrix(y_test, y_pred_knn)
print("Confusion Matrix:\n", con_mat_knn)
```

```python
# precision
prec_knn = con_mat_knn[0][0]*100/(con_mat_knn[0][0] + con_mat_knn[0][1])
print("Precision:", prec_knn)
```

```python
# recall
recall_knn = con_mat_knn[0][0]*100/(con_mat_knn[0][0] + con_mat_knn[1][0])
print('Recall:', recall_knn)
```

```python
# F-1 Score
f1_knn=2*prec_knn*recall_knn/(recall_knn + prec_knn)
print("F1 Score:", f1_knn)
```

Conclusion: The model finally can predict 92.22% times the correct accuracy if the person will happen to take credit card on not. While 7.78% times the model is unable to predict. We have achieved a better precision with high accuracy thus model can be used for future aspirations. Although accuracy is higher, a more better hypertuning can further increase the accuracy.
